```python
teams=teams.dropna()
```

```python
teams
```

```python
train=teams[teams['year']<2012].copy()
test=teams[teams['year']>=2012].copy()
```

```python
train.shape
```

```python
test.shape
```

```python
from sklearn. linear_model import LinearRegression
reg= LinearRegression()
```

```python
predictores=['athletes','prev_medals']
target='medals'
```

```python
reg.fit(train[predictores],train['medals'])
```

```python
predictions=reg.predict(test[predictores])
predictions
```

```python
test['predictores']=predictions
```

```python
test
```

```python
test.loc[test['predictores']<0,'predictores']=0
test
```

```python
test['predictores']=test['predictores'].round()
test
```

```python
from sklearn.metrics import mean_absolute_error
error=mean_absolute_error(test['medals'],test['predictores'])
error
```

```python
errors=(test['medals']-test['predictores']).abs()
errors
```

```python
error_by_team=errors.groupby(test['team']).mean()
error_by_team
```

```python
medals_by_team=test['medals'].groupby(test['team']).mean()
medals_by_team
```

```python
error_ratio=error_by_team/medals_by_team
error_ratio
```
